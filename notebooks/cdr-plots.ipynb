```python
from casacore.tables import table
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = table('/safepool/gh/dsa110/mar2022/newrefant.G0')
```

```python
scan_no = t.getcol("SCAN_NUMBER")
```

```python
scan_no.shape
```

```python
ant1 = t.getcol("ANTENNA1")
cparam = t.getcol('CPARAM')
```

```python
gain = cparam[:,0,0]
```

```python
flags = ~t.getcol('FLAG')
```

```python
flag = flags[:,0,0]
```

```python
gain.shape
```

```python
scan_no_f = scan_no[flag]
gain_f = gain[flag]
ant1_f = ant1[flag]
```

```python
to_plot = gain_f[ant1_f==20]
plt.plot(np.angle(to_plot, deg=True))
```

```python
ant1.shape
```

```python
np.unique(ant1_f).shape
```

```python
plt.figure(figsize=(8,6))
plt.rcParams['font.size'] = '16'
time = np.arange(8) * 5
for a in np.unique(ant1_f)[::8]:
    to_plot = gain_f[ant1_f==a]
    plt.plot(time, np.angle(to_plot, deg=True))
plt.xlabel('Time (min)')
plt.ylabel('Gain phase (Deg)')
plt.title("DSA-110 gain phase over time")
```

```python
plt.figure(figsize=(8,6))
time = np.arange(8) * 5
for a in ant1_f:
    to_plot = gain_f[ant1_f==a]
    plt.plot(time, np.abs(to_plot))
plt.xlabel('Time (min)')
plt.ylabel('Gain amp (Deg)')
```
