```python
from plotpsf import get_psf_stats
```

```python
import numpy as np
```

```python
dist, radial_average, radial_std, radial_maxabs = get_psf_stats(
    '/safepool/data/AJ-15int-4000chan/15x60s-4000chan-briggs-0-75-psf.fits', 2)
```

```python
from matplotlib import pyplot as plt
import matplotlib
```

```python
plt.figure(figsize=(8,5))
plt.plot(dist, radial_average, label='avgabs')
plt.plot(dist, radial_std, label='std')
plt.plot(dist, radial_maxabs, label='maxabs')
plt.xlabel('Offset(arcmin)')
plt.ylabel('PSF')
plt.yscale('log')
plt.grid()
plt.ylim([1e-6, 1])
plt.title('Briggs(0.75), 3.06asec x 3.05asec')
plt.legend()

matplotlib.rcParams.update({'font.size': 16})
```

```python
matplotlib.rcParams.update({'font.size': 16})
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(18,5))
for ax in (ax1, ax2):
    ax.plot(dist, radial_average, label='avgabs')
    ax.plot(dist, radial_std, label='std')
    ax.plot(dist, radial_maxabs, label='maxabs')
    ax.set_yscale('log')
    ax.grid()
    ax.set_ylim([1e-6, 1])
    ax.legend()
    ax.set_xlabel('Offset(arcmin)')
ax1.set_ylabel('PSF')
ax1.set_xlim([0, 7.5])
plt.suptitle('AJ Briggs(0.75), 3.06asec x 3.05asec')

```
