```python
from casacore.tables import table
import numpy as np
import matplotlib.pyplot as plt
```

```python
with table('/fastpool/data/W-1chan.ms') as t:
    weights = t.getcol('IMAGING_WEIGHT_SPECTRUM')
    uvw = t.getcol('UVW')
    weights = np.zeros_like(weights)
    uvw=np.zeros_like(uvw)
    t.getcolnp('IMAGING_WEIGHT_SPECTRUM', weights)
    t.getcolnp('UVW', uvw)
```

```python
plt.plot(uvw[:,0], uvw[:,1], linestyle="None", marker='o', markerfacecolor='None', markeredgecolor='C0', markersize=0.1, alpha=0.1)
# plt.scatter(-uvw[::10,0], -uvw[::10,1], c=weights[::10,0,0], s=0.1)
#plt.colorbar()
plt.xlabel('m')
plt.ylabel('m')
plt.gca().set_aspect('equal')
```

```python
from matplotlib.ticker import MultipleLocator
fig, ax = plt.subplots()
ax.plot(uvw[:,0], uvw[:,1], linestyle="None", marker='o', markerfacecolor='C0', markeredgecolor='C0', markersize=5)
ax.set_xlim([13000, 13200])
ax.set_ylim([400, 600])
ax.set_aspect('equal')
ax.yaxis.set_minor_locator(MultipleLocator(5))
ax.xaxis.set_minor_locator(MultipleLocator(5))
ax.grid(which='both', color='#CCCCCC')
plt.xlabel('m')
plt.ylabel('m')
```
